# Answer-1
We have given two matrices A and B, one of size A(n, k) and the other one is of size B(k, m).

After multiplying these matrices we get a matrix of size C(n,m).
$$ C=A*B$$
Total elements that we have in matrix C is    nm

Now during matrix multiplication we perform some operations, 


and every element gets multiply k times during matrix multiplication and undergoes k-1 time addition operation.


So Number of multiplication we need to perform is:$$ nmk$$
So Number of addition we need to perform is:$$ n*m*k-1$$


# Answer-2

```python
#Ans-2 
#Multiplication of matrices
import timeit

def matrix_multiplication(matrix1,matrix2):
    result_matrix=[]
    for i in range(len(matrix1)):
        matrix=[]
        for j in range(len(matrix2[0])):
            matrix.append(0)
        result_matrix.append(matrix)
        
    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                result_matrix[i][j]+=matrix1[i][k]*matrix2[k][j]
    return result_matrix

matrix1=[[1,2],[2,3]]
matrix2=[[2,4,3],[2,1,5]]

time_taken = timeit.timeit(lambda: matrix_multiplication(matrix1, matrix2), number=10000)
print("Time taken is",(time_taken*1000)/10000,'ms')
multiplication=matrix_multiplication(matrix1,matrix2)
print("Multiplication of matrix without using numpy\n",multiplication)


            
    
```

```python
#Verification

import numpy as np

time_take = timeit.timeit(lambda: np.dot(matrix1,matrix2), number=10000)
print("Time taken is",(time_take*1000)/10000,'ms')
multiplication_new=np.dot(matrix1,matrix2)
print("Multiplication of matrix using numpy\n",multiplication_new)
        
```

The solution which we get using numpy is faster.
The reason is the algorithm which we wrote in order of order O(n^3) which takes larger time to execute. While numpy is implemented in C so the algorithm for multiplication is more optimised for such operations.

In order to find the median using the given method we have iterate through the complete array n/2 times. So time complexity in this case would be $$ n^{\frac{n}{2}} $$
A better method to find the median of the array could be:
In a sorted array median is either the middle element or the average to middle elements.



```python
import numpy as np
import timeit
np.random.seed(0)
def median(arr):
    n=len(arr)
    sort_arr=sorted(arr, reverse=True)
    #If array is of even length means it will have two middle elements
    if n%2==0:
        mid1=sort_arr[n//2-1]
        mid2=sort_arr[n//2]
        middle=(mid1+mid2)/2
    else:
        middle=sort_arr[n//2]

arr=np.random.randint(1,1000,size=(1000))
time_take = timeit.timeit(lambda: median(arr), number=10000)
print("Time taken by proposed algorithm is",(time_take*1000)/10000,'ms')

#using numpy
np.random.seed(0)
arr1=np.random.randint(1,1000,size=(1000))
time_take = timeit.timeit(lambda: np.median(arr1), number=10000)
print("Time taken by numpy median is",(time_take*1000)/10000,'ms')


        
```

# Answer-4 
We have   $$f(x,y)= x^2y+y^3sin(x)
$$
Now for gradient of function f(x,y) with respect x is:
$$\frac{\partial f(x,y)}{\partial x}=2xy+y^3cos(x)$$

Now for gradient of function f(x,y) with respect y is:
$$\frac{\partial f(x,y)}{\partial y}=x^2+3y^2sin(x)$$

   

# Answer-5

```python
#Ans 5
import jax
import jax.numpy as jnp

def given_function(x,y):
    return x**2*y+y**3*jnp.sin(x)
gradient=jax.grad(given_function,argnums=(0,1))

x_val = [1.0,2.0,3.0]
y_val = [2.0,1.0,4.0]
print(" Gradient using Jax")
for x,y in zip(x_val,y_val):
    print('x=',x,'y=',y)
    print("x gradient",gradient(x,y)[0])
    print("y gradient",gradient(x,y)[1],"\n")
    
# For analytical solution and verification

def analytical(x,y):
    gradient_x=2*x*y+y**3*jnp.cos(x)
    gradient_y=x**2+3*y**2*jnp.sin(x)
    return gradient_x,gradient_y

print(" \nGradient using Analytical solution\n")
for x,y in zip(x_val,y_val):
    print('x=',x,'y=',y)
    print("x gradient",analytical(x,y)[0])
    print("y gradient",analytical(x,y)[1],"\n")

    
```

From the above results we can see that Jax solution matches with Analytical solution

```python
#Ans-8
# As per the given information we can make three classes, 1. years  2. Branch  3.student

class year:
    def __init__(self,year):
        self.year=year
        self.branches=[]
    def add_branches(self,branch_name):
        self.branches.append(branch_name)

class branches:
    def __init__(self,branch):
        self.branch=branch
        self.student=[]
    def add_student(self,student):
        self.student.append(student)

class student:
    def __init__(self,roll_no,name,marks):
        self.roll_no=roll_no
        self.name=name
        self.marks=marks
#creating objects.

student1=student(1,'N',{'maths':100,'English':70})
branch1=branches('Branch1')
branch1.add_student(student1)
branch2=branches('Branch2')

        

    
#Year objects
year1=year(2022)
year1.add_branches(branch1)
year1.add_branches(branch2)

branch_year2_1=branches('Branch1')
branch_year2_2=branches('Branch2')

year2=year(2023)
year2.add_branches(branch_year2_1)
year2.add_branches(branch_year2_2)

branch_year3_1=branches('Branch1')
branch_year3_2=branches('Branch2')

year3=year(2024)
year3.add_branches(branch_year3_1)
year3.add_branches(branch_year3_2)

branch_year4_1=branches('Branch1')
branch_year4_2=branches('Branch2')

year4=year(2025)
year4.add_branches(branch_year4_1)
year4.add_branches(branch_year4_2)

#Overall database is list of year objects
database=[year1,year2,year3,year4]

#Printing data for branch 1 of 2022
print(database[0].branches[0].student[0].roll_no)
print(database[0].branches[0].student[0].name)
print(database[0].branches[0].student[0].marks['maths'])
print(database[0].branches[0].student[0].marks['English'])

    
```

```python
#Ans-13
def argmin(array):
    index=0
    min_value=array[0]
    for i in range(1,len(arr)):
        if arr[i]<min_value:
            min_value=arr[i]
            index=i
    return index
        
arr=np.array([2,3,5,1,4])
index_argmin=argmin(arr)
print("index using argmin function",index_argmin)

# Verification
index=np.argmin(arr)
print("\nVerification\n")
print("index using numpy argmin function",index)


    
```
